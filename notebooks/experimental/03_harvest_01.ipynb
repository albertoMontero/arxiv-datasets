##### Playing with csv

```python
import csv
```

```python
chunk = chunk1[0]
type(chunk)
```

```python
fieldnames = list(chunk.keys())
fieldnames
```

```python
output_file = "data/chunks.csv"
```

```python
with open(output_file, 'w', newline="\n") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(chunk1)
```
